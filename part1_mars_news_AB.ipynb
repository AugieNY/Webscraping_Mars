```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Mars NASA news site
url = 'https://redplanetscience.com'
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html = browser.html
news_soup = soup(html, 'html.parser')
```

```python
# Extract all the text elements
news_soup.div
```

```python
# Create an empty list to store the dictionaries
summary = []
```

```python
#Extract the titles and preview text of the news articles that you scraped. 
#Store the scraping results in Python data structures as follows:
#Store each title-and-preview pair in a Python dictionary. 
#And, give each dictionary two keys: title and preview. 

#titles
doc_title =news_soup.find('div', class_='content_title')
```

```python
#previews
doc_preview = news_soup.find_all('div', class_='article_teaser_body')
```

```python
# length / number or articles per page 
len(doc_preview)
```

```python
zl = zip(doc_title,doc_preview)
```

```python
#Store all the dictionaries in a Python list.
# Print the list in your notebook.
zl = list(zl)
print(zl)
```

```python
#test your code

test =[]
for i,n in zl:
    inter = {}
    
    number = i.text
    
    abstract = n.text.strip('\u200b')
    
    inter = {
        'doc_title':number,
        'doc_preview': abstract
    }
    
    test.append(inter)
    
test
```

```python
# Loop through the text elements
# Extract the title and preview text from the elements
# Store each title and preview pair in a dictionary
# Add the dictionary to the list

for i in range(15):
    
    dic = {} 
    
    title = news_soup.find_all('div', class_='content_title')[i].text
    
    preview = news_soup.find_all('div', class_='article_teaser_body')[i].text.strip('\u200b')
    
    dic = {
        'title': title,
        'preview': preview
    }
    
    summary.append(dic)
```

```python
# Print the list to confirm success
summary
```
